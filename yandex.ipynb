```python
from random import randint
import time
```

```python
%matplotlib notebook
```

```python
n = 10
L1 = [randint(1,n) for i in range(n)]
L2 = [randint(1,n) for i in range(3)]
```

```python
print('Первый список :',L1)
print('Второй список :',L2)
print('Результат     :',[el for el in L1 if el not in L2])
print('Результат ч/з множество, теряется последовательность :', set(L1) - set(L2))
```

# Оценка О() для списка и для множества

```python
time_value_list = []
time_value_set = []
L2 = [randint(1,n) for i in range(100)]
for deg in range(4,9):

    L1 = [randint(5,5**deg) for i in range(5**deg)]
    t0 = time.time()
    [el for el in L1 if el not in L2]
    time_value_list.append((time.time()-t0)*1000)
    
    t0 = time.time()
    set(L1) - set(L2)
    time_value_set.append((time.time()-t0)*1000)
```

# Оценка

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x =np.arange(4,9)
y = np.array(time_value_list)
print(y)
```

```python
plt.figure("список list")
plt.plot(x, y, '-o')
```

```python
y = np.array(time_value_set)
print(y)
```

```python
plt.figure("список set")
plt.plot(x, y, '-o')
```
